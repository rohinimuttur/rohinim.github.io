```python
import itertools
query_bu="select distinct BU from Common.BrandMaster"
df_bu=pd.read_sql(query_bu,cnxn)
bu_sub_list=df_bu.values.tolist()
bu_list = list(itertools.chain(*bu_sub_list))
bu_list

query_sub_bu="select distinct SubBU from Common.BrandMaster"
df_subbu=pd.read_sql(query_sub_bu,cnxn)
subbu_sub_list=df_subbu.values.tolist()
subbu_list = list(itertools.chain(*subbu_sub_list))

query_pepsi_brand="select distinct PepsiBrand from Common.BrandMaster"
df_pepbrnd=pd.read_sql(query_pepsi_brand,cnxn)
pepbrnd_sub_list=df_pepbrnd.values.tolist()
pep_brnd_list = list(itertools.chain(*pepbrnd_sub_list))

query_brand_src="select distinct Source from Common.BrandMaster"
df_brnd_src=pd.read_sql(query_brand_src,cnxn)
brndsrc_sub_list=df_brnd_src.values.tolist()
brnd_src_list = list(itertools.chain(*brndsrc_sub_list))

```

```python
with open('bu.txt', 'w') as fh:  
    for i in bu_list:
        fh.write('%s\n' % i)
        
with open('subbu.txt', 'w') as fh:  
    for i in subbu_list:
        fh.write('%s\n' % i)
        
with open('pepsibrand.txt', 'w') as fh:  
    for i in pep_brnd_list:
        fh.write('%s\n' % i)

with open('brndsrc.txt', 'w') as fh:  
    for i in brnd_src_list:
        fh.write('%s\n' % i)
```
